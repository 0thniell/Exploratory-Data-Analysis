```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None
```

```python
df = pd.read_excel(r'C:\Users\IT SUPPORT\Desktop\financial_ratio.xlsx')

print(df.head())
```

Descriptive Statistics


```python
descriptive_stats = df.describe()
print(descriptive_stats)
```

Descriptive Statistics by Category

```python
liquidity_ratios = df[['Current_ratio', 'Quick_ratio', 'Cash_ratio']]
efficiency_ratios = df[['Inventory_Turnover_Ratio', 'Accounts_Receivable_Turnover_Ratio', 'Assets_Turnover_Ratio']]
profitability_ratios = df[['Profit_margin', 'Return_on_Assets', 'Return_on_Equity']]
solvency_ratios = df[['Debt_to_Assets_Ratio', 'Debt_to_Equity_Ratio', 'Interest_Coverage_Ratio']]

liquidity_stats = liquidity_ratios.describe()
efficiency_stats = efficiency_ratios.describe()
profitability_stats = profitability_ratios.describe()
solvency_stats = solvency_ratios.describe()


print("Liquidity Ratios:")
print(liquidity_stats)
print("\n")

print("Efficiency Ratios:")
print(efficiency_stats)
print("\n")

print("Profitability Ratios:")
print(profitability_stats)
print("\n")

print("Solvency Ratios:")
print(solvency_stats)
```

Correlation Analysis


```python
correlation_matrix = df.corr()
print(correlation_matrix)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Financial Ratios')
plt.show()

```
